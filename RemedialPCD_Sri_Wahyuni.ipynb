```python
# Import pustaka yang dibutuhkan
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Membaca gambar menggunakan OpenCV
image_path = '/content/WhatsApp Image 2024-12-03 at 12.32.13.jpeg'  # Ganti dengan path gambar Anda
image = cv2.imread(image_path)

# Konversi gambar dari BGR ke RGB untuk visualisasi Matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 2. Konversi gambar ke ruang warna HSV
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# 3. Mendefinisikan rentang warna merah dalam ruang warna HSV
# Rentang untuk warna merah (dua rentang karena merah berada di ujung spektrum)
lower_red1 = np.array([0, 120, 70])
upper_red1 = np.array([10, 255, 255])
lower_red2 = np.array([170, 120, 70])
upper_red2 = np.array([180, 255, 255])

# Masking warna merah pada gambar
mask1 = cv2.inRange(image_hsv, lower_red1, upper_red1)
mask2 = cv2.inRange(image_hsv, lower_red2, upper_red2)
red_mask = mask1 + mask2

# 4. Menggunakan mask untuk menampilkan hanya area merah pada gambar
red_detected = cv2.bitwise_and(image_rgb, image_rgb, mask=red_mask)

# Visualisasi hasil
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.title("Gambar Asli")
plt.imshow(image_rgb)
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Mask Warna Merah")
plt.imshow(red_mask, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Deteksi Warna Merah")
plt.imshow(red_detected)
plt.axis("off")

plt.tight_layout()
plt.show()
```
